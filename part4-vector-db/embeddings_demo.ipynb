```python
from sentence_transformers import SentenceTransformer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
# Cricket
"Virat Kohli scored a century",
"The bowler took a wicket",
"India won the match",

# Cooking
"I cooked pasta today",
"The recipe needs salt",
"Baking requires precision",

# Cybersecurity
"Phishing attacks are common",
"Use strong passwords",
"Firewall protects systems",

"Cyber attacks are increasing"
]

embeddings = model.encode(sentences)

# Similarity matrix
similarity = np.inner(embeddings, embeddings)

sns.heatmap(similarity)
plt.show()

# Query
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

scores = np.inner(query_embedding, embeddings)[0]

top_idx = scores.argsort()[-2:][::-1]

for i in top_idx:
    print(sentences[i], scores[i])
```
